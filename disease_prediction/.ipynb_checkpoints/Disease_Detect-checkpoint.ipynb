```python
## selecting K best
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',)

k_best=SelectKBest(score_func=chi2,k=40)

print(x_train.shape)
k_best.fit(x_train,y_train)

# X_train_new = SelectKBest(chi2, k=40).fit(x_train,y_train)
# imp_indices = SelectKBest(chi2, k=40).get_support(indices=True

print(df.columns[k_best.get_support(indices=True)])
x_train = k_best.fit_transform(x_train,y_train)
print(x_train.shape)
```

```python
kbest_features=list(df.columns[k_best.get_support(indices=True)])
type(list(df.columns[k_best.get_support(indices=True)]))
```

```python
dt = dt.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
test_df=pd.read_csv('Testing.csv')
test_df = test_df.sample(frac = 1)

x_test = test_df.drop(['prognosis'],axis =1)
x_test = x_test[kbest_features].copy()
print(x_test.shape)
y_test = test_df['prognosis']

y_pred=dt.predict(x_test)

accuracy_score(y_test,y_pred)
```

```python
real_diseases = y_test.values
print(real_diseases)
print(y_pred)
```

```python
for i in range(0, len(real_diseases)):
    if y_pred[i] == real_diseases[i]:
        print ('{0}:: Pred: {1} Actual:{2}'.format(i+1,y_pred[i], real_diseases[i]))
    else:
        print('worng prediction')
        print ('{0}:: Pred: {1} Actual:{2}'.format(i+1,y_pred[i], real_diseases[i]))
```
